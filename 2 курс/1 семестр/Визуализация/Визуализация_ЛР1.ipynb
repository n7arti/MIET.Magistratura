```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp

# Интервал от 0 до 20 включительно
x = np.linspace(0, 20, 1000)

# Функции Бесселя
y1 = sp.jv(0, x/2)  # Функция Бесселя y = J0(x/2)
y2 = sp.jv(3, x)  # Функция Бесселя y = J3(x)

# Создание графика
plt.figure(figsize=(10, 6))

# Настройка фона окна отображения графика
plt.gcf().set_facecolor('white')  # Фон всей фигуры
plt.gca().set_facecolor('white')  # Фон рабочей области

# Построение графика функции Бесселя J0(x/2)
plt.plot(x, y1, label='Функция Бесселя J0(x/2)', color='purple', linewidth=2)

# Построение графика функции Бесселя J3(x)
plt.plot(x, y2, label='Функция Бесселя J3(x)', color='red', linewidth=2)

# Настройка осей и сетки
plt.xlabel('x', fontsize=14, color='black')
plt.ylabel('y', fontsize=14, color='black')
plt.title('Графики функций Бесселя', fontsize=16, color='blue')

# Установка числа меток на осях и числа линий сетки
plt.xticks(np.arange(0, 21, 2), fontsize=12, color='black')  # Метки по оси X от 0 до 20 с шагом 2
plt.yticks(np.arange(-0.5, 1.2, 0.2), fontsize=12, color='black')  # Метки по оси Y от -0.5 до 1.1 с шагом 0.2
plt.grid(True, which='both', linestyle='--', color='black')  # Включаем сетку для обеих осей

# Дополнительные надписи на графике
plt.annotate(r'$J_0(x/2)$', xy=(2, sp.jv(0, 2/2)), xytext=(3, 0.75),
             textcoords='data', arrowprops=dict(facecolor='darkblue', shrink=0.05, width=2),
             fontsize=12, color='darkblue')

plt.annotate(r'$J_3(x)$', xy=(18, sp.jv(3, 18)), xytext=(18, 0.35),
             textcoords='data', arrowprops=dict(facecolor='darkblue', shrink=0.05, width=2),
             fontsize=12, color='darkblue')

# Отображение графика
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Интервал для X и Y
x = np.linspace(-10, 10, 1000)
y = np.linspace(-2, 2, 1000)
x, y = np.meshgrid(x, y)

# Функции для построения поверхности
z = np.cos(x)*np.exp(-y**2)

# Создание фигуры и 3D оси
fig = plt.figure(figsize=(7, 10))

# Первый график в стиле jet
ax1 = fig.add_subplot(211, projection='3d')
surf1 = ax1.plot_surface(x, y, z, cmap='jet')
ax1.set_title('jet', fontsize=16, color='black')
ax1.set_xlabel('X', fontsize=12)
ax1.set_ylabel('Y', fontsize=12)
ax1.set_zlabel('Z', fontsize=12, labelpad=1)
fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=10)

# Второй график в стиле cool
ax2 = fig.add_subplot(212, projection='3d')
surf2 = ax2.plot_surface(x, y, z, cmap='cool')
ax2.set_title('cool', fontsize=16, color='black')
ax2.set_xlabel('X', fontsize=12)
ax2.set_ylabel('Y', fontsize=12)
ax2.set_zlabel('Z', fontsize=12, labelpad=1)
fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=10)

# Поворот графиков
ax1.view_init(elev=15, azim=50)
ax2.view_init(elev=15, azim=50)

# Настройка белого фона рабочей области
ax1.set_facecolor('white')
ax2.set_facecolor('white')

# Установка числа меток на осях и числа линий сетки
ax1.set_zticks(np.arange(-1, 1.1, 0.5))
ax1.set_yticks(np.arange(-2, 2.1, 1))
ax1.set_xticks(np.arange(-10, 10.1, 5))

ax2.set_zticks(np.arange(-1, 1.1, 0.5))
ax2.set_yticks(np.arange(-2, 2.1, 1))
ax2.set_xticks(np.arange(-10, 10.1, 5))

# Отображение графиков
plt.show()
```
