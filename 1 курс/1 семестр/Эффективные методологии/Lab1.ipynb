````python
x = np.linspace(0,10)
y1 = 4*x
y2 = [i**2 for i in x]
plt.title("Графики зависимостей: y1 = 4x, y2 = x^2")
plt.xlabel("x")
plt.ylabel("y1,y2")
plt.grid(which='both')
plt.xticks(np.arange(0, 11, 0.5), fontsize=8, color='gray')
plt.yticks(np.arange(0, 110, 5), fontsize=8, color='gray')

plt.plot(x, y2, label='y = x^2')
plt.scatter(x,y1, label='y = 4x', color = "red")
plt.legend()
````

````python
import scipy.optimize as opt
from scipy.optimize import fsolve
import numpy as np
import math

second = 1000000
minute = 60*second
hour = 60*minute
day = 24*hour
month = 30*day
year = 12*month
century = 100*year

def equation_to_solve1(n):
    return n * np.log(n) - century


a = math.sqrt(century)
# Поиск корня уравнения с использованием метода бисекции
b = opt.bisect(equation_to_solve1, 1, century)
c = math.log(century, 2)
n = 0
while math.factorial(n) < century:
    n += 1

print("n =", n-1)
````

````python
n = np.linspace(1, 100)
f1 = [np.log(i) for i in n]
f2 = [math.sqrt(i) for i in n]
f3 = n
f4 = [i*np.log(i) for i in n]
f5 = [i**2 for i in n]
f6 = [i**3 for i in n]
f7 = [2**int(i) for i in n]
f8 = [math.factorial(int(i)) for i in n]

plt.subplot(2,4,1)
plt.plot(n, f1)
plt.grid()
plt.title("ln n")

plt.subplot(2,4,2)
plt.plot(n, f2)
plt.grid()
plt.title("sqrt(n)")

plt.subplot(2,4,3)
plt.plot(n, f3)
plt.grid()
plt.title("n")

plt.subplot(2,4,4)
plt.plot(n, f4)
plt.grid()
plt.title("n*ln n")

plt.subplot(2,4,5)
plt.plot(n, f5)
plt.grid()
plt.title("n^2")

plt.subplot(2,4,6)
plt.plot(n, f6)
plt.grid()
plt.title("n^3")

plt.subplot(2,4,7)
plt.plot(n, f7)
plt.grid()
plt.title("2^n")

plt.subplot(2,4,8)
plt.plot(n, f8)
plt.grid()
plt.title("n!")
````

````python
n = np.linspace(2, 100, 99)
f1 = [np.log(i) for i in n]
f2 = [math.sqrt(i) for i in n]
f3 = n
f4 = [i*np.log(i) for i in n]
f5 = [i**2 for i in n]
f6 = [i**3 for i in n]
f7 = [2**int(i) for i in n]
f8 = [math.factorial(int(i)) for i in n]

plt.title("Графики функций")
plt.loglog(n, f1, label='f(n) = ln n', color = "red")

plt.loglog(n, f2,label='f(n) = sqrt(n)', color = "blue")

plt.loglog(n, f3, label='f(n) = n', color = "green")


plt.loglog(n, f4, label='f(n) = n*ln n', color = "black")

plt.loglog(n, f5, label='f(n) = n^2', color = "yellow")

plt.loglog(n, f6, label='f(n) = n^3', color = "pink")

plt.loglog(n, f7, label='f(n) = 2^n', color = "purple")

#plt.plot(n, f8, label='f(n) = n!', color = "gray")
plt.legend()
plt.grid()
plt.ylabel("f(n)", fontsize=14)
plt.xlabel("n",fontsize=14)
````
