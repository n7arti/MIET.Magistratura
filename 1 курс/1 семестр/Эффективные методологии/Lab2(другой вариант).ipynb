```python
def bubble_sort_recursive(arr, n):
    if n == 1:
        return
    for i in range(n-1):
        if arr[i] > arr[i+1]:
            arr[i], arr[i+1] = arr[i+1], arr[i]
    bubble_sort_recursive(arr, n-1)

# Пример использования
arr = [64, 34, 25, 12, 22, 11, 90]
n = len(arr)
bubble_sort_recursive(arr, n)
print("Отсортированный массив:")
for i in range(n):
    print("%d" % arr[i], end=" ")
```

**Инвариант цикла**: После каждой итерации внутреннего цикла самый большой элемент "всплывает" на правильную позицию в массиве.<br>
**Инициализация**: Перед началом цикла массив неотсортирован. <br>
**Сохранение**: После каждой итерации внутреннего цикла, самый большой элемент "всплывает" на правильную позицию, что поддерживает инвариант. <br>
**Завершение**: После завершения цикла весь массив упорядочен по возрастанию.<br>
Цикл завершается, когда не остаётся ни одной пары элементов, которые нуждаются в обмене, что обеспечивает завершение алгоритма.

# Методом анализа алгоритма
Внутренний цикл совершает n−i−1 сравнений
Внешний цикл выполняется n−1 раз, так как после каждого прохода наибольший элемент становится на своё место.Таким образом, общее количество сравнений и обменов можно оценить как сумму арифметической прогрессии:
(n−1)+(n−2)+…+1= n(n−1)/2 = (n<sup>2</sup>−n)/2
Отсюда мы можем заключить, что количество операций, выполняемых в процессе пузырьковой сортировки, имеет квадратичную зависимость от размера входного массива. Следовательно, асимптотическая сложность пузырьковой сортировки в худшем и среднем случае составляет O(n<sup>2</sup>).<br>
# Методом подстановки
Предположим, что время выполнения может быть представлено как <br>
T(n) = aT(n/b) + O(nᶜ)<br>
Предположим, что размер массива равен 7.<br>
T(n) = 1T(n/6) + O(n) <br>
Предположим, что T(m)≤C⋅mᵈ для всех  m < n, где C и d - некоторые константы.<br>
Используя это предположение, давайте попробуем доказать, что T(n)≤C⋅nᵈ
Подставим наше предположение:<br>
T(n/6)≤C * (n/6)ᵈ<br>
Теперь заменим это обратно в исходное уравнение:<br>
T(n/6)≤C * (n/6)ᵈ + O(n) <br>
Упростим это: <br>C/6ᵈ * nᵈ + O(n)<br>

Таким образом, сравнивая это с нашим первоначальным предположением, мы можем увидеть, что T(n)≤C⋅nᵈ при условии, что C≥ C/6ᵈ. Таким образом, асимптотическая сложность данного рекуррентного соотношения составляет О(n)
Асимптотическая сложность алгоритма О(n)
# Методом деревьев рекурсии<br>
T(n) = 1T(n/6) + O(n) <br>
На 1 уровне: O(n)<br>
На 2 уровне: O((n/6)<sup>2</sup>) = 1/36n<sup>2</sup><br>
На 2 уровне: O((n/6)<sup>3</sup>) = 1/216n<sup>3</sup><br>
...<br>
На последнем О(1)<br>
Общая сложность алгоритма будет равна О(n + 1/36n<sup>2</sup> + 1/216n<sup>3</sup> + ...) что можно упростить до O(n) <br>
